# Testing Hardcover API with GraphQL

```python
import requests
import pandas as pd
import os
```

```python
query = """
query {
  me {
    username
    books_count
    lists(limit: 50) {
      books_count
      list_books {
        user_books {
          user_book_reads {
            finished_at
          }
        }
      }
    }
  }
}
"""
url = "https://api.hardcover.app/v1/graphql"

headers = {
    "authorization": f"Bearer {os.environ['TOKEN']}",
    "content-type": "application/json"
}
```

```python
response = requests.post(url, json={'query': query}, headers=headers)

data = response.json()
data
```
