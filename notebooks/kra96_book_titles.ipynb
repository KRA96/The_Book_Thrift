# Create Book ID and Title Dataset

```python
import pandas as pd
import os
import time
from IPython.display import clear_output
```

```python
# Read books data chunk
books = pd.read_json("../raw_data/goodreads_books.json.gz",
                     compression="gzip",
                     lines=True,
                     nrows=10_000)
```

```python
[(i, col) for i, col in enumerate(books.columns.tolist())]
```

```python
book_titles = books.iloc[:, [0, 18, 24, 26, 27]]
book_titles.dtypes, book_titles.info()
```

```python
book_titles.loc[book_titles["work_id"] == '', "work_id"] = pd.NA
```

```python
book_titles = book_titles.astype({
    "isbn": pd.StringDtype(),
    "isbn13": pd.StringDtype(),
    "book_id": pd.Int32Dtype(),
    "work_id": pd.Int32Dtype(),
    "title": pd.StringDtype()
})
```

```python
# Reading and adding chunk by chunk
for chunk_no, chunk in enumerate(pd.read_json("../raw_data/goodreads_books.json.gz",
                          compression="gzip",
                          lines=True,
                          chunksize=50_000)):
    clear_output(wait=True)
    start = time.time()
    # use only columns needed
    chunk = chunk.iloc[:, [0, 18, 24, 26, 27]]

    # Fill empty workid with na
    chunk.loc[chunk["work_id"] == '', "work_id"] = pd.NA

    # Downcast dtypes
    chunk = chunk.astype({
    "isbn": pd.StringDtype(),
    "isbn13": pd.StringDtype(),
    "book_id": pd.Int32Dtype(),
    "work_id": pd.Int32Dtype(),
    "title": pd.StringDtype()
    })
    # Write to csv
    csv_path = "../raw_data/book_titles.csv"
    chunk.to_csv(csv_path, mode="a", header=not os.path.exists(csv_path))
    print(f"added chunk no {chunk_no}")

end = time.time()
print(f"Time taken: {end - start}")
```
