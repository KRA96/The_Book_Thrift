```python
from bookread_pipeline import build_pipeline, load_raw, make_cluster_labels
import pandas as pd
```

```python
df = pd.read_csv("./raw_data/goodreads_interactions.csv", nrows=50_000)
```

```python
from interaction_pipeline import build_interactions_pipeline

pipe = build_interactions_pipeline()

pipe.fit(df)
df_processed = pipe.transform(df)
```

```python
M = pipe.named_steps["matrix"].matrix_   # CSR user x book
enc = pipe.named_steps["encode"]
```

```python
from sklearn.preprocessing import normalize
import numpy as np

def most_similar_book_cosine(book_id, pipe, M):
    enc = pipe.named_steps["encode"]
    b_idx = enc.book_to_index_.get(book_id)
    if b_idx is None:
        raise ValueError("book_id inconnu.")


    M_items_norm = normalize(M.T, norm="l2")

    v = M_items_norm[b_idx].T
    sims = M_items_norm.dot(v).toarray().ravel()

    sims[b_idx] = -np.inf
    best_idx = int(np.argmax(sims))

    return enc.index_to_book_[best_idx], float(sims[best_idx])
```

```python
best_book_id, score = most_similar_book_cosine(12345, pipe, M)
print(best_book_id, score)
```
